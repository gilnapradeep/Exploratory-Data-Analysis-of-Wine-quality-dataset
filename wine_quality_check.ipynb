## Exploratory Data Analysis of Wine Quality Dataset

## 1. Introduction

Wine quality evaluation is an important task in the wine industry. The taste and quality of wine depend on several chemical properties such as acidity, alcohol content, sugar levels, and sulphates.

The Wine Quality Dataset contains physicochemical attributes of wine samples and their quality ratings given by experts.

## Objective

The objective of this project is to:

 - Perform exploratory data analysis (EDA) on the dataset

 - Understand relationships between chemical features and wine quality

 - Identify important features affecting wine quality

 - Detect patterns, trends, and outliers in the data

## 2. Dataset Description

The dataset consists of 1143 wine samples and 12 attributes.
- Input Variables : fixed acidity, volatile acidity, citric acid, residual sugar, chlorides, free sulfur dioxide, total sulfur dioxide, density, pH, sulphates, alcohol.
- Target Variables : quality. 

## 3. Import Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 4. Loading the Dataset

```python
df = pd.read_csv('WineQT.csv', index_col="Id")
```

```python
df.isnull().sum()
```

The dataset does not contain missing values, so no data cleaning is required.

## 7. Distribution of Target Variable

```python
sns.countplot(x = 'quality', data=df)
plt.title('Distribution of Wine Quality')
plt.show()
```

Observation

Most wines have quality ratings of 5 or 6, indicating that the dataset is slightly imbalanced toward medium-quality wines.

## 8. Univariate Analysis

```python
df.hist(figsize=(12,10))
plt.show()
```

Observations

 - Alcohol values are mostly between 9% and 11%

 - Residual sugar values are concentrated at lower levels

 - Sulphates show some outliers

## 9. Correlation Analysis

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot = True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Observations

 - Alcohol shows a positive correlation with quality

 - Volatile acidity shows a negative correlation with quality

 - Density negatively correlates with alcohol

## 10. Relationship Between Quality and Features

```python
sns.boxplot(x='quality', y='alcohol', data=df)
plt.title('Alcohol Content vs Quality')
plt.show()
```

Observation:

Higher quality wines tend to have higher alcohol content.

```python
sns.violinplot(x='quality', y='volatile acidity', data=df)
plt.title('Volatile Acidity vs Quality')
plt.show()
```

Observation:

Lower volatile acidity corresponds to better wine quality.

## 11. Pairwise Feature Relationships

```python
sns.pairplot(df[['alcohol','volatile acidity','citric acid','sulphates','quality']], hue='quality')
```

Observation:

The plot helps visualize relationships among important features and their influence on quality.

## 12. Feature Impact on Quality

```python
df.groupby('quality').mean().transpose().plot(figsize=(12,8))
plt.title("Average Feature Values by Quality")
plt.show()
```

Observation:

This plot shows how average chemical properties change as wine quality increases.

## 13. Outlier Detection

```python
df.boxplot(figsize=(20,15))
plt.show()
```

Observation:

Outliers are present in:

 - Free sulfur dioxide

 - Total sulfur dioxide

 - Fixed acidity

## 14. Key Findings

From the exploratory analysis, the following insights were observed:

 1. Most wines are rated between 5 and 6.

 2. Alcohol content has a strong positive relationship with wine quality.

 3. Volatile acidity negatively impacts wine quality.

 4. Sulphates and citric acid show moderate positive correlations with quality.

 5. Several features contain outliers which may influence model performance.

## 15. Conclusion

In this project, exploratory data analysis was performed on the Wine Quality dataset. Various statistical and visualization techniques were used to understand relationships between physicochemical properties and wine quality.

The analysis revealed that alcohol content, volatile acidity, and sulphates are among the most influential factors affecting wine quality.

These insights can be further used to build machine learning models for predicting wine quality.
